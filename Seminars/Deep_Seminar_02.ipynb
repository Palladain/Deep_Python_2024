## Простое наследование

````python
class Bird:
    def __init__(self):
        print("Вызвался класс bird")


class Parrot(Bird):
    pass


first_class = Bird()
second_class = Parrot()
````

Базовый класс, или родительские классы, - это классы, от которых наследуются данные или методы. Подкласс - в данном примере `Parrot` - наследуется от суперкласса. Также считается, что любой подкласс является производным от своего родительского класса или подкласс расширяет родительский класс.

````python
class Bird:

    name = 'Ted'

    def __init__(self):
        print("Вызвался класс bird")


class Parrot(Bird):
    pass


first_class = Bird()
second_class = Parrot()

print(second_class.name)
````

Тут мы можем заметить, что объекты базового класса становятся общими для всех дочерних классов

### Наследование от множества классов

Множественное наследование - одна из ключевых особенностей наследования.

В принципе, все просто: это возможность класса иметь более одного родительского класса. При множественном наследовании дочерний класс наследует все свойства родительских классов. Однако, используя множественное наследование, следует быть предельно внимательными.

Давайте разберемся с множественным наследованием на примере класса утки, бобра и утконоса.



````python
# Класс Утка
class Duck:
    def __init__(self):
        self.species = "Утка"

    def swim(self):
        """Метод для плавания"""
        print(f"{self.species} плывет по воде.")

    def quack(self):
        """Метод для кряканья"""
        print(f"{self.species} крякает: Кря-кря!")

# Класс Бобер
class Beaver:
    def __init__(self):
        self.species = "Бобер"

    def build_dam(self):
        """Метод для строительства плотин"""
        print(f"{self.species} строит плотину из веток.")

    def chew_wood(self):
        """Метод для жевания деревьев"""
        print(f"{self.species} грызет дерево своими зубами.")

# Класс Утконос, наследующийся от Утки и Бобра
class Platypus(Duck, Beaver):
    def __init__(self):
        super().__init__()
        self.species = "Утконос"

    def lay_eggs(self):
        """Метод для откладывания яиц (свойство утконоса)"""
        print(f"{self.species} откладывает яйца, хотя это млекопитающее!")

    def display_abilities(self):
        """Метод, демонстрирующий все способности утконоса"""
        self.swim()
        self.quack()
        self.build_dam()
        self.chew_wood()
        self.lay_eggs()

# Пример использования
platypus = Platypus()
platypus.display_abilities()
````

Ого получается два наследования, а какой порядок?

На этот вопрос отвечает MRO! Это порядок, в котором Python ищет методы и атрибуты при вызове метода в классе с множественным наследованием. В нашем примере класс Platypus наследует методы сначала от Duck, а затем от Beaver.

При запуске программы MRO для Platypus выведет следующий порядок:


````python
print("MRO:", [cls.__name__ for cls in type(platypus).mro()])
````

### Mixin

Окей, мы поговорили про то, что можем наследоваться от двух классов. Но мы забыли поговорить про такую вещь, как `mixin`.

`mixin` (или примесь, или миксин) – это дополнительный класс, методы которого предназначены для использования в других классах.

Давайте разберемся на примере робота-пылесоса.

У нас есть базовые классы – пылесос и машина на радиоуправлении.

От пылесоса нам достается базовый метод, что наш робот всасывает предметы, а от машины на радиоуправлении у нас есть возможность ездить. Но этого недостаточно,чтобы сделать автономного робота. И давайте добавим дополнительно класс, который позволит нам автономно двигаться по площади уборки

Семинары - Frame 2.jpg

````python
# Базовый класс для пылесоса
class VacuumCleaner:
    def __init__(self) -> None:
        self.dust_collected = 0  # Собранная пыль

    def vacuum(self):
        """Метод для всасывания пыли"""
        self.dust_collected += 1
        print(f"Впитывание пыли... Собрано {self.dust_collected} единиц пыли.")

# Базовый класс для машины на радиоуправлении
class RemoteControlCar:
    def __init__(self) -> None:
        self.position = [0, 0]  # Начальная позиция
        self.direction = "N"  # Направление: N, E, S, W

    def move_forward(self, distance: int) -> None:
        """Метод для движения вперед"""
        if self.direction == "N":
            self.position[1] += distance
        elif self.direction == "E":
            self.position[0] += distance
        elif self.direction == "S":
            self.position[1] -= distance
        elif self.direction == "W":
            self.position[0] -= distance
        print(f"Двигаемся вперед на {distance} единиц. Позиция: {self.position}")

    def move_backward(self, distance: int) -> None:
        """Метод для движения назад"""
        if self.direction == "N":
            self.position[1] -= distance
        elif self.direction == "E":
            self.position[0] -= distance
        elif self.direction == "S":
            self.position[1] += distance
        elif self.direction == "W":
            self.position[0] += distance
        print(f"Двигаемся назад на {distance} единиц. Позиция: {self.position}")

    def turn_left(self) -> None:
        """Поворот налево"""
        directions = ["N", "W", "S", "E"]
        self.direction = directions[(directions.index(self.direction) + 1) % 4]
        print(f"Поворот налево. Теперь направление: {self.direction}")

    def turn_right(self) -> None:
        """Поворот направо"""
        directions = ["N", "E", "S", "W"]
        self.direction = directions[(directions.index(self.direction) + 1) % 4]
        print(f"Поворот направо. Теперь направление: {self.direction}")

# Класс для автономного движения
class AutonomousMovement:
    def __init__(self) -> None:
        self.sensors = {"front": False, "left": False, "right": False}  # Ложь означает, что препятствий нет

    def detect_obstacle(self, direction) -> dict:
        """Метод для распознавания препятствия"""
        return self.sensors[direction]

    def auto_move(self) -> None:
        """Метод для автономного движения"""
        if not self.detect_obstacle("front"):
            print("Препятствий впереди нет, едем вперед.")
            self.move_forward(1)
        else:
            print("Обнаружено препятствие! Пытаемся объехать...")
            if not self.detect_obstacle("left"):
                self.turn_left()
                self.move_forward(1)
            elif not self.detect_obstacle("right"):
                self.turn_right()
                self.move_forward(1)
            else:
                print("Заблокирован со всех сторон. Остановка.")

# Итоговый класс автономного робота для уборки
class AutonomousCleaningRobot(VacuumCleaner, RemoteControlCar, AutonomousMovement):
    def __init__(self) -> None:
        VacuumCleaner.__init__(self)
        RemoteControlCar.__init__(self)
        AutonomousMovement.__init__(self)

    def clean_and_move(self) -> None:
        """Метод для автономной уборки и движения"""
        print("Начинаем уборку...")
        self.vacuum()  # Включаем всасывание
        self.auto_move()  # Автономное движение

# Пример использования
robot = AutonomousCleaningRobot()

# Пылесосим и двигаемся
robot.clean_and_move()

# Изменение состояния датчиков для проверки автономного движения
robot.sensors["front"] = True
robot.clean_and_move()
````

### Ромбовидное наследование

Давайте рассмотрим ситуацию, когда наше наследование выглядит таким образом:

Семинары - Frame 3.jpg

````python
class BaseClass:
    num_base_calls = 0
    def call_me(self) -> None:
        print("Calling method on BaseClass" )
        self.num_base_calls += 1


class LeftSubClass(BaseClass):
    num_left_calls = 0
    def call_me(self) -> None:
        BaseClass.call_me(self)
        print("Calling method on LeftSubClass")
        self.num_left_calls += 1


class RightSubClass(BaseClass):
    num_right_calls = 0
    def call_me(self) -> None:
        BaseClass.call_me(self)
        print("Calling method on RightSubClass")
        self.num_right_calls += 1


class Subclass(LeftSubClass, RightSubClass):
    num_sub_calls = 0
    def call_me(self) -> None:
        LeftSubClass.call_me(self)
        RightSubClass.call_me(self)
        print("Calling method on Subclass")
        self.num_sub_calls += 1
````

````python
s = Subclass()

s.call_me()

print(f" s.num_sub_calls: {s.num_sub_calls} \n",
      f"s.num_left_calls: {s.num_left_calls} \n",
      f"s.num_right_calls: {s.num_right_calls} \n",
      f"s.num_base_calls: {s.num_base_calls} \n")
````

Как мы можем заметить, тут вызывается два раза BaseClass. В нашем случае может быть это не так критично, но вы представьте, если мы так будем подключаться к  тяжелой базе данных, например, два и более раз. Давайте подумаем, как можно решить данную проблему

````python
Subclass.mro()
````

````python
# your code:

class BaseClass:
    num_base_calls = 0
    def call_me(self) -> None:
        print("Calling method on BaseClass" )
        self.num_base_calls += 1


class LeftSubClass(BaseClass):
    num_left_calls = 0
    def call_me(self) -> None:
        super().call_me()
        print("Calling method on LeftSubClass")
        self.num_left_calls += 1


class RightSubClass(BaseClass):
    num_right_calls = 0
    def call_me(self) -> None:
        super().call_me()
        print("Calling method on RightSubClass")
        self.num_right_calls += 1


class Subclass(LeftSubClass, RightSubClass):
    num_sub_calls = 0
    def call_me(self) -> None:
        super().call_me()
        print("Calling method on Subclass")
        self.num_sub_calls += 1
````

````python
s = Subclass()

s.call_me()

print(f" s.num_sub_calls: {s.num_sub_calls} \n",
      f"s.num_left_calls: {s.num_left_calls} \n",
      f"s.num_right_calls: {s.num_right_calls} \n",
      f"s.num_base_calls: {s.num_base_calls} \n")
````

````python
Subclass.mro()
````

Используя порядок разрешения методов (MRO, Method Resolution Order), преобразуем ромб в плоский линейный кортеж. Результаты этого можно про­ анализировать в атрибуте __mro_ класса. Линейная версия данного ромба представляет собой последовательность SubClass, LeftSubClass, RightSubClass, BaseClass, object. Здесь важно то, что SubClass определяет LeftSubClass перед RightSubClass, применяя порядок к классам в ромбе.
При множественном наследовании нужно помнить, что часто оказывается необходимо вызывать следующий метод в последовательности MRO, и это не обязательно метод родительского класса. Функция `super()` находит имя в последовательности MRO. Действительно, функция `super()` изначально была создана для того, чтобы сделать возможными сложные формы множественного наследования.


Если простыми словами, то `super()` ведет нас от нулевого по индексу класса направо, тем самым не давая нам дублировать вызовы классов.

## Полиморфизм

В лекции мы узнали, что такое полиморфизм.



> Полиморфизм - возможность реализовать один и тот же метод разными способами (например, для жизни одни попугаи ищут дупло, другие строят гнездо)


Так давайте реализуем пример полиморфизма.

````python
### Your code:

# Базовый класс
class Bird:
    def build_nest(self):
        raise NotImplementedError("This method should be overridden in subclasses")

# Дочерний класс Parrot1 (ищет дупло)
class Parrot1(Bird):
    def build_nest(self):
        return "Parrot1 ищет дупло для гнезда"

# Дочерний класс Parrot2 (строит гнездо)
class Parrot2(Bird):
    def build_nest(self):
        return "Parrot2 строит гнездо на дереве"

# Пример использования
birds = [Parrot1(), Parrot2()]

for bird in birds:
    print(bird.build_nest())
````

Да, так просто реализуется принцип полиморфизма

## Абстракция

Давайте рассмотрим пример с нейронными сетями.

````python
class BaseClassNN:

    def __init__(self):
        print("init neural network")

    def train(self):
        print("Model is training")


class CNN(BaseClassNN):
    def __init__(self):
        super().__init__()

    def predict(self):
        print("predicted dog on the picture")


class VGG(BaseClassNN):
    def __init__(self):
        super().__init__()

    def train(self):
        super().train()
````

````python
vgg = VGG()

vgg.train()

vgg.predict()
````

Как мы видим, мы забыли прописать метод предсказания.
И чтобы такой ошибки не возникало, надо в базовом классе прописать метод `predict`.
Это и будет абстракция, которая позволит обходить ошибки.

````python
class BaseClassNN:

    def __init__(self):
        print("init neural network")

    def train(self):
        print("Model is training")

    def predict(self):
        raise NotImplementedError("This method should be overridden in subclasses")


class CNN(BaseClassNN):
    def __init__(self):
        super().__init__()

    def predict(self):
        print("predicted dog on the picture")


class VGG(BaseClassNN):
    def __init__(self):
        super().__init__()

    def train(self):
        super().train()
````

Pydantic - это библиотека Python для разбора и валидации данных, которая использует аннотации типов Python. Она особенно полезна в ситуациях, где важны валидация данных, их преобразование и обеспечение соответствия типам. Хотя Python является динамически типизированным языком, Pydantic предлагает надежный способ убедиться, что данные, с которыми вы работаете, соответствуют определенным ограничениям и форматам.

Использование:

* Разработка API

Pydantic часто используется с FastAPI для стандартизации конечных точек API (увидим дальше)
* Загрузка данных

При работе с внешними источниками данных (например, базами данных, CSV-файлами или сторонними API) Pydantic помогает обеспечить хорошую структурированность входящих данных и их соответствие ожидаемым форматам / обеспечить миграцию структур этих данных (увидим на следующих семинарах).
* Управление конфигурацией

Модели Pydantic идеально подходят для управления настройками и конфигурациями приложений, обеспечивая соблюдение типов и значений по умолчанию. (сейчас посмотрим)


````python
from pydantic import BaseModel, ValidationError

class UserModel(BaseModel):
    id: int
    name: str
    email: str

# Valid Input
try:
    user = UserModel(id=1, name='John Doe', email='john@example.com')
    print(user)
except ValidationError as e:
    print(e)

# Invalid Input
try:
    invalid_user = UserModel(id='not_a_number', name='Jane Doe', email='jane@example.com')
except ValidationError as e:
    print(e)
````

````python
from enum import Enum
from typing import Union, Optional
from pydantic import BaseModel
import openai
from openai import OpenAI

client = OpenAI()


class ModifyEventActionType(Enum):
    CREATE = "create"
    MODIFY = "modify"
    CANCEL = "cancel"

class GetEventsByDateFCParams(BaseModel):
    """Returns all calendar events on a given date or on a range of dates. Range starts at date_from and ends at date_to"""

    date_from: str | None = Field(
        default=None,
        description="Optional date in yyyy-MM-dd format from (including) which to search. It will be set to current date by default. If presented, `date_to` should also be presented",
    )
    date_to: str = Field(
        description="Date in yyyy-MM-dd format up to which in the future to perform search, including this date"
    )

class EventDuration(BaseModel):
    """Event duration represented as sum of days (if lasting more that 1 day), hours and minutes."""

    days: Optional[int] = Field(
        description="If event is lasting more than one day, this is used to set the duration in days"
    )
    hours: Optional[int] = Field(description="Hours part of event duration 0-23")
    minutes: Optional[int] = Field(description="Minutes part of event duration 0-59")


class ModifyEventFCParams(BaseModel):
    """Use this when you need to create new event/reminder or modify an existing event / reminder.
    Fill those fields you can infer from the context, leaving the rest blank.
    For modifying an event, you need to know its id. Use your search tools to find it out if you don't have it at hand.
    When modifying, you can only predict those fields that change"""

    action: ModifyEventActionType = Field(description="Designates the action you are taking.")
    event_id: str | None = Field(
        description="Id of event. Only required when modyfying an existing event."
    )
    title: str | None = Field(default=None, description="Title of the event being created or modified")
    description: str | None = Field(
        description="1-sentence description of the event being created or modified. Fill it with the information about how this event should be communicated (if applicable)",
    )
    event_date: str | None = Field(
        description="If inferable from context: date of the event, or the first event in the series, if it is recurring. Should be in yyyy-MM-dd format, i.e. 2024-05-25",
    )
    event_time: str | None = Field(
        description="If inferable from context: time the event starts. Should be in HH:mm:ss or HH:mm format, i.e. 18:30:00 or 18:30",
    )
    duration: EventDuration | None = Field(
        description="Event duration as a combination of days, hours and minutes"
    )
````

````python
tools = [openai.pydantic_function_tool(GetEventsByDateFCParams), openai.pydantic_function_tool(ModifyEventFCParams)]

messages = []
messages.append({"role": "system", "content": "You are a helpful customer support assistant. Use the supplied tools to assist the user."})
messages.append({"role": "user", "content": "Hi, can you please help me to create a reminder about HW on python course on 23/09 23:59?"})
````

````python
response = client.chat.completions.create(
    model='gpt-4o-2024-05-13',
    messages=messages,
    tools=tools
)
````

````python
class CalendarEvent(BaseModel):
    title: str | None = Field(default=None, description="Title of the event being created or modified")
    description: str | None = Field(
        description="1-sentence description of the event being created or modified. Fill it with the information about how this event should be communicated (if applicable)",
    )
    event_date: str | None = Field(
        description="If inferable from context: date of the event, or the first event in the series, if it is recurring. Should be in yyyy-MM-dd format, i.e. 2024-05-25",
    )
    event_time: str | None = Field(
        description="If inferable from context: time the event starts. Should be in HH:mm:ss or HH:mm format, i.e. 18:30:00 or 18:30",
    )
    duration: EventDuration | None = Field(
        description="Event duration as a combination of days, hours and minutes"
    )
````

````python
completion = client.beta.chat.completions.parse(
    model="gpt-4o-2024-08-06",
    messages=[
        {"role": "system", "content": "Extract the event information."},
        {"role": "user", "content": "Alice and Bob are going to a science fair on Friday."},
    ],
    response_format=CalendarEvent,
)

event = completion.choices[0].message.parsed
````
